```python
import requests as req 
from bs4 import BeautifulSoup
import pandas as pd
import json
import re
# from pprint im
```

```python
S_url= 'https://www.royalcanin.com/es/dogs/breeds/breed-library?sizes=a5e42826-cbbf-466a-b121-fcbc11e83425&filter=dcac5bcc-83a0-4a96-93fe-7ae986867f1b&viewAll=true'
```

```python
S_html = req.get(S_url).content
```

```python

S_soup = BeautifulSoup(S_html, 'html.parser')
S_soup
```

```python


S_dogs = S_soup.find_all('h3')

S_name_list= []

for tag in S_dogs: 
   S_name_list.append(tag.text)

S_name_list= S_name_list[:-1]
S_name_list
```

```python
div_tags = S_soup.find_all('div', {'class': 'rc-column'})

S_url_list = []

for div_tag in div_tags:
    a_tags = div_tag.find_all('a', {'class': 'rc-card__link rc-full-width', 'href': True})
    for a_tag in a_tags:
        href = a_tag.get('href')
        S_url_list.append(href)

S_url_list
```

```python

image_tags = S_soup.find_all('img')

image_urls = []

for img_tag in image_tags:
    if 'data-srcset' in img_tag.attrs:
        srcset = img_tag['data-srcset']
        urls = srcset.split(',')
        for url in urls:
            url = url.strip().split(' ')[0]
            if 'w=320' in url and 'nutrition' not in url and 'library' not in url:
                image_urls.append(url)

image_urls
```

```python
import requests
from bs4 import BeautifulSoup

div_tags = S_soup.find_all('div', {'class': 'rc-column'})

href_list = []
div_list = []

for div_tag in div_tags:
    a_tags = div_tag.find_all('a', {'class': 'rc-card__link rc-full-width', 'href': True})
    for a_tag in a_tags:
        href = a_tag.get('href')
        
        # Scrape the corresponding <div class="rc-column">
        response = requests.get(href)
        soup = BeautifulSoup(response.content, 'html.parser')
        target_div = soup.find('div', {'class': 'rc-padding--sm rc-margin-bottom--xs rc-bg-colour--brand3'})
        
        # Process the scraped <div> element as needed
        if target_div:
            div_elements = target_div.find_all('div', {'class': 'rc-column'})
            div_texts = [div.text.strip() for div in div_elements]
            
            href_list.append(href)
            div_list.append(div_texts)
        else:
            href_list.append(href)
            div_list.append(None)

# Print the scraped data
print(href_list)
print(div_list)
```

```python
pd.set_option('display.max_colwidth', None)

S_Dogsdf= pd.DataFrame(div_list)
S_Dogsdf
```

```python
pd.set_option('display.max_colwidth', None)


# Replace all occurrences of \n with a space in the first column


# Assuming you have the DataFrame S_Dogsdf and you want to modify the first column (column index 0)
S_Dogsdf.iloc[:, 2] = S_Dogsdf.iloc[:, 2].str.replace('Informacin clave', '')
# Replace "\n\n\n" with a line change
S_Dogsdf.iloc[:, 2] = S_Dogsdf.iloc[:, 2].str.replace('\n\n\n', '\n')
S_Dogsdf.iloc[:, 2] = S_Dogsdf.iloc[:, 2].str.replace('\n', ' - ')

S_Dogsdf
```

```python
pd.set_option('display.max_colwidth', None)

S_Dogsdf.iloc[:, 1] = S_Dogsdf.iloc[:, 1].str.replace('Especificidades de la raza\n\n', '')
S_Dogsdf.iloc[:, 1] = S_Dogsdf.iloc[:, 1].str.replace('\n\n', ' Personalidad: ')
S_Dogsdf.iloc[:, 1] = S_Dogsdf.iloc[:, 1].str.replace('Pas\n', 'Pas: ')
S_Dogsdf.iloc[:, 1] = S_Dogsdf.iloc[:, 1].str.replace('Categora de tamao\n', 'Categora de tamao: ')
S_Dogsdf.iloc[:, 1] = S_Dogsdf.iloc[:, 1].str.replace('Esperanza de vida promedio\n', 'Esperanza de vida promedio: ')
S_Dogsdf.iloc[:, 1] = S_Dogsdf.iloc[:, 1].str.replace('\n', ' ')

S_Dogsdf
```

```python
S_Dogsdf.rename(columns= {0: 'Acerca de', 1 : 'Especificidades de la raza', 2 : 'Informacin clave'}, inplace=True)

S_Dogsdf
```
