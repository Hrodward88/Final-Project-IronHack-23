```python
import requests as req 
from bs4 import BeautifulSoup
import pandas as pd
import json
import re
import numpy as np
# from pprint im
```

```python
Alergia_url= 'https://www.trusens.com/es-mx/news/blog/best-hypoallergenic-dogs/'
```

```python
Alergia_html = req.get(Alergia_url).content
```

```python

Alergia_soup = BeautifulSoup(Alergia_html, 'html.parser')
Alergia_soup
```

```python


Alergia_dogs = Alergia_soup.find_all('h2')

Alergia_name_list= []

for tag in Alergia_dogs: 
   Alergia_name_list.append(tag.text)

Alergia_name_list= Alergia_name_list[:-1]
Alergia_name_list.sort()
Alergia_name_list

```

```python
Alergia_name_list = [name.replace('\xa0', '') for name in Alergia_name_list]
```

```python
Alergia2_url= 'https://www.redcanina.es/razas-de-perros-para-duenos-con-alergia/'
Alergia2_html = req.get(Alergia2_url).content
Alergia2_soup = BeautifulSoup(Alergia2_html, 'html.parser')
Alergia2_soup

Alergia2_dogs = Alergia2_soup.find_all('h3')

Alergia2_name_list= []

for tag in Alergia2_dogs: 
   Alergia2_name_list.append(tag.text)

Alergia2_name_list= Alergia2_name_list[:-1]
Alergia2_name_list
Alergia2_name_list = Alergia2_name_list[:21]
Alergia2_name_list
```

```python
Alergia3_url= 'https://www.businessinsider.es/10-razas-perros-personas-alergicas-ellos-811729'
Alergia3_html = req.get(Alergia3_url).content
Alergia3_soup = BeautifulSoup(Alergia3_html, 'html.parser')
Alergia3_soup

Alergia3_dogs = Alergia3_soup.find_all('h2')

Alergia3_name_list= []

for tag in Alergia3_dogs: 
   Alergia3_name_list.append(tag.text)

Alergia3_name_list= Alergia3_name_list[:-1]
Alergia3_name_list = Alergia3_name_list[:21]
Alergia3_name_list
```

```python
Alergia4_url= 'https://www.micasarevista.com/mascotas-perros-gatos/g42118456/razas-perros-hipoalergenicos-no-sueltan-pelo-alergias/'
Alergia4_html = req.get(Alergia4_url).content
Alergia4_soup = BeautifulSoup(Alergia4_html, 'html.parser')
Alergia4_soup

Alergia4_dogs = Alergia4_soup.find_all('h2', {'class': 'css-1dhct3x e1tmud0h7'})

Alergia4_name_list= []

for tag in Alergia4_dogs: 
   Alergia4_name_list.append(tag.text)

Alergia4_name_list= Alergia4_name_list[:-1]
Alergia4_name_list.sort()
Alergia4_name_list
```

```python
Alergia = pd.DataFrame(Alergia_name_list)
Alergia
```

```python
# Create a new DataFrame from Alergia2_name_list
Alergia2 = pd.DataFrame({'Alergia2_name_list': Alergia2_name_list})

# Concatenate the two DataFrames
Alergia = pd.concat([Alergia, Alergia2], axis=1)

Alergia3 = pd.DataFrame({'Alergia3_name_list': Alergia3_name_list})

Alergia = pd.concat([Alergia, Alergia3], axis=1)

Alergia4 = pd.DataFrame({'Alergia4_name_list': Alergia4_name_list})

Alergia = pd.concat([Alergia, Alergia4], axis=1)

Alergia

```
