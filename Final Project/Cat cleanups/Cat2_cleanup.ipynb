```python
import requests as req 
from bs4 import BeautifulSoup
import pandas as pd
import json
import re
# from pprint im
```

```python
S_url= 'https://www.purina.es/encuentra-mascota/razas-de-gato?page=%2C1'
```

```python
S_html = req.get(S_url).content
```

```python

S_soup = BeautifulSoup(S_html, 'html.parser')
S_soup
```

```python


S_dogs = S_soup.find_all('h4')

S_name_list= []

for tag in S_dogs: 
   S_name_list.append(tag.text)

S_name_list= S_name_list[:-1]
S_name_list
```

```python
breed_names = []

for name in S_name_list:
    # Remove leading and trailing whitespaces
    name = name.strip()
    # Remove leading and trailing newline characters
    name = name.strip('\n')
    breed_names.append(name)

# Print the breed names
print(breed_names)
```

```python
div_tags = S_soup.find_all('h4', {'class': 'results-view-name'})

S_url_list = []

for div_tag in div_tags:
    a_tags = div_tag.find_all('a', {'href': True})
    for a_tag in a_tags:
        href = a_tag.get('href')
        complete_url = 'https://purina.es'+ href
        S_url_list.append(complete_url)

S_url_list
```

```python
image_tags = S_soup.find_all('img', {'loading': 'lazy', 'class': 'image-style-nppe-breed-selector-500'})

image_urls = []

for img_tag in image_tags:
    if 'src' in img_tag.attrs:
        src = img_tag['src']
        full_url = 'https://purina.es'+ src
        image_urls.append(full_url)

image_urls
```

```python
import requests
from bs4 import BeautifulSoup

div_tags = S_soup.find_all('h4', {'class': 'results-view-name'})

href_list = []
div_texts_list = []
col_elements_list = []
summary_list = []

for div_tag in div_tags:
    a_tags = div_tag.find_all('a', {'href': True})
    for a_tag in a_tags:
        href = a_tag.get('href')
        complete_url = 'https://purina.es'+ href
        S_url_list.append(complete_url)

        # Scrape the corresponding <div class="rc-column">
        response = req.get(complete_url)
        soup = BeautifulSoup(response.content, 'html.parser')

        # Scrape the <div class="hero-small--content-area-intro">
        intro_div = soup.find('div', {'class': 'hero-small--content-area-intro'})
        if intro_div:
            intro_text = intro_div.text.strip()
        else:
            intro_text = None

        # Scrape the <div class="row">
        row_div = soup.find('div', {'class': 'row'})
        if row_div:
            col_elements = row_div.find_all('div', {'class': 'col-md-4'})
            col_texts = [col.text.strip() for col in col_elements]
        else:
            col_elements = None
            col_texts = None

        # Scrape the second <div class="clearfix text-formatted field field--name-field-c-sideimagetext-summary field--type-text-long field--label-hidden field__item">
        summary_divs = soup.find_all('div', {'class': 'clearfix text-formatted field field--name-field-c-sideimagetext-summary field--type-text-long field--label-hidden field__item'})
        if len(summary_divs) >= 2:
            summary_text = summary_divs[1].text.strip()
        else:
            summary_text = None

        href_list.append(complete_url)
        div_texts_list.append(intro_text) 
        col_elements_list.append(col_texts)
        summary_list.append(summary_text)

# Print the scraped data
print(href_list)
print(div_texts_list)
print(col_elements_list)
print(summary_list)
```
